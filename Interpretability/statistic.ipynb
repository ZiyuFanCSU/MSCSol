## Molecular weight 

```python
import matplotlib.pyplot as plt
import numpy as np
import json
with open('/dataset/benchmark.json', 'r') as f:
    data = json.load(f)      

dataset = data
mw = [pred["MW"] for pred in data]

# Create the histogram with density=True to show probability density instead of counts
plt.figure(figsize=(3, 2)) 
plt.hist(mw, bins=6, color='grey', alpha=0.7, density=False, rwidth=0.9)  # density=True makes the y-axis represent probability density

# Adding titles and labels
plt.title('Histogram of Molecular Weights', fontsize=14)
plt.xlabel('Molecular Weight', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Remove the box around the plot
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.ylim(0, 3500)
# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data=np.random.rand(100)
plt.figure(figsize=(3, 2)) 
plt.hist(data,bins=10,edgecolor='black',cumulative=False, density=False)
hist,bin_edges=np.histogram(data,bins=10)
cumulative_freg =np.cumsum(hist)

ax = plt.gca()
ax2 = ax.twinx()
ax2.plot(bin_edges[1:], cumulative_freg, 'r-')
ax2.set_ylabel('Cumulative Frequency')
plt.title("Histogram with Cumulative Frequency Distribution")
plt.xlabel("Value")
plt.ylabel("Cumulative Frequency")
plt.show()
```
