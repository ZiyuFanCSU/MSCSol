## TSNE

```python
import matplotlib.pyplot as plt
import json
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

train_x_fea = np.loadtxt('./pic/gvp_re_feature.csv', delimiter=',')
y_test = []
with open('./pic/gvp_re_solubility.txt', 'r') as f:
    for line in f:
        y_test.append(line.strip())

tsne = TSNE(n_components=2, init='pca', random_state=6)
X_pca = tsne.fit_transform(np.array(train_x_fea))

import numpy as np
import pandas as pd

x = []
y = []
for i in X_pca:
    x.append(i[0])
    y.append(i[1])
plt.scatter(x, y, c=[float(val) for val in y_test], cmap='PuOr',s=2,alpha=0.5) 
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

train_x_fea = np.loadtxt('./pic/gvp_cl_feature.csv', delimiter=',')
y_test = []
with open('./pic/gvp_cl_solubility.txt', 'r') as f:
    for line in f:
        y_test.append(line.strip())
y = []
for i in y_test:
    if float(i) >= -1:
        y.append(0)
    elif float(i) >= -3:
        y.append(1)
    else:
        y.append(2)

import seaborn as sns  
from sklearn.manifold import TSNE

def pic(train_x_fea,y,t):
    X_pca = t.fit_transform(np.array(train_x_fea))
    
    print(X_pca.shape)
    high_x, high_y = [], []
    mid_x, mid_y = [], []
    low_x, low_y = [], []
    for i in range(len(X_pca)):
        if y[i] == 0:
            low_x.append(X_pca[i][0])
            low_y.append(X_pca[i][1])
        elif y[i] == 1:
            mid_x.append(X_pca[i][0])
            mid_y.append(X_pca[i][1])
        else:
            high_x.append(X_pca[i][0])
            high_y.append(X_pca[i][1])

    sns.set(font_scale=1.2)
    plt.scatter(low_x, low_y, c='#dd621f', marker='o',s=2,alpha=0.8)
    plt.scatter(mid_x, mid_y, c='#f5f7cb', marker='o',s=3,alpha=0.8)
    plt.scatter(high_x, high_y, c='#5793ca', marker='o',s=2,alpha=0.8)
    plt.show()

tsne = TSNE(n_components=2, init='pca', random_state=6)
pic(train_x_fea,y,tsne)
```
